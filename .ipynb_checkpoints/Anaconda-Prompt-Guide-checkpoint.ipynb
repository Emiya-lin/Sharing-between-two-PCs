# Anaconda Prompt Guide
参考[Anaconda](https://docs.conda.io/projects/conda/en/latest/user-guide/index.html)教程网站

## 常用命令
---
获取版本号：
> conda --version \
> conda -V

获取帮助信息：
> conda --help \
> conda -h

查看某一命令的帮助信息如update命令或remove命令
> conda update --help \
> conda remove --help

上述命令中的`--help`也可以替换为`-h`.

### 管理环境

查看环境管理的全部命令：
> conda env -h

创建新环境：
> conda create -n py3.7-gpu python=3.7

上述命令创建了Python版本为3.7的环境，并命名为py3.7-gpu.

切换环境：
> activate py3.7-gpu (this command for Windows) \
> source activate py3.7-gpu (this command for Linux) \
> conda activate py3.7-gpu (this command also for Linux)

回到base环境（无需添加base环境名）：
> conda deactivate (this command for Windows) \
> source conda deactivate (this command for Linux)

列出所有环境：
>conda info --envs \
>conda env list

二者起到相同作用。

删除环境env_name：
> conda env remove -n env_name \
> conda remove -n env_name

貌似加不加env都可以。

### 管理包

列举当前活跃环境下的所有包(package):
> conda list

列举非当前活跃环境的所有包：
> conda list -n your_env_name

下载包：
> conda install package_name

如果遇到无法下载最新版本软件包的问题，比如node.js最新版本已经有15.0.0，但是清华镜像站只有13.0.0，此时可以开启`VPN`然后用`conda`官方推荐的channel下载：
> conda install package_name -c conda-forge

## 重命名环境：
---
先创建新环境并把旧环境的内容复制过去：
> conda create -n new_env --clone old_env

第二步，删除旧环境的所有文件：
>conda remove -n old_env --all

## 分享环境：
---
如果你想把你当前的环境配置与别人分享，这样ta可以快速建立一个与你一模一样的环境来共同开发新的试验。一个分享环境的快速方法就是给ta一个你环境的`.yml`文件。

首先通过`activate target_env`切换到你要分享的环境`target_env`，然后输入以下命令在当前工作目录先生成一个`environment.yml`文件：
> conda env export > environment.yml

小伙伴拿到`environment.yml`文件后，将该文件放在工作目录下，通过以下命令从该文件创建环境：
> conda env create -f environment.yml
